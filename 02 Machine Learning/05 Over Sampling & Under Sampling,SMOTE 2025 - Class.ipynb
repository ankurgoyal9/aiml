```python
import pandas as pd
import numpy as np
import imblearn   #conda install -c conda-forge imbalanced-learn
import seaborn as sns
from sklearn.datasets import make_classification # genrates data for u
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
```

```python
# defining the dataset
X, y = make_classification(n_samples= 10000, weights=[.99]) # creating 10k rows with 99 percent imbalance data
# weight to make it balance n inbalance
```

```python
y.shape
```

```python
# class distribution
print(Counter(y))
```

```python
#Random Oversampling
# instantiating the random oversampler 
ros = RandomOverSampler()
# resampling X, y
X_ros, y_ros = ros.fit_resample(X, y)
```

```python
print(np.unique(y_ros,return_counts=True))
```

```python
sns.scatterplot(x=X_ros[:,0],y=X_ros[:,1],hue=y_ros,alpha=.5)
```

```python
#Random Undersampling

# instantiating the random undersampler
rus = RandomUnderSampler() 
# resampling X, y
X_rus, y_rus = rus.fit_resample(X, y)
```

```python
print(np.unique(y_rus,return_counts=True))
```

```python
sns.scatterplot(x=X_rus[:,0],y=X_rus[:,1],hue=y_rus)
```

# So above results are enough

```python
# instantiating over and under sampler
over = RandomOverSampler(sampling_strategy=0.5)
under = RandomUnderSampler(sampling_strategy=0.8)
```

```python
# first performing oversampling to minority class
X_over, y_over = over.fit_resample(X, y)
print(f"Oversampled: {Counter(y_over)}") # f =====> format
```

```python
# now to comine under sampling 
X_combined_sampling, y_combined_sampling = under.fit_resample(X_over, y_over) 
# using above X_over, y_over 
print(f"Combined Random Sampling: {Counter(y_combined_sampling)}") # 
```
