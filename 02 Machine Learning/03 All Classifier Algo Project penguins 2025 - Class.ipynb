Handelling the missing values

**Spliting independent and dependent**

```python
ind=data.drop(columns="species")
dep=data.species
```

```python
ind.head()
```

```python
ind=pd.get_dummies(ind,dtype=int,drop_first=True)
```

```python
ind.head()
```

```python
ind.shape,dep.shape
```

#### Spliting data

```python
from sklearn.model_selection import train_test_split
```

```python
x,xt,y,yt=train_test_split(ind,
                          dep,
                          test_size=.2,
                          random_state=42,
                          stratify=data.species)  # stratify maintain the tartget ratio same as parent for classes
```

```python
x.shape,xt.shape
```

```python
y.value_counts(normalize=True),yt.value_counts(normalize=True)
```
