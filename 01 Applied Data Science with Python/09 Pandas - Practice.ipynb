```python
import pandas as pd

# Create a sample dataframe
df = pd.DataFrame({
    'fruit': ['apple', 'banana', 'apple', 'orange', 'banana'],
    'size': ['small', 'large', 'medium', 'large', 'small'],
    'price': [1.0, 0.8, 1.2, 1.5, 0.9]
})

# Convert the 'fruit' column to dummy variables
fruit_dummies = pd.get_dummies(df['fruit'], prefix='fruit')

# Display the original dataframe and the dummy variables
print("Original dataframe:")
print(df)
print("\nDummy variables for fruit column:")
print(fruit_dummies)

# You can also convert multiple columns at once
all_dummies = pd.get_dummies(df[['fruit', 'size']])
print("\nDummy variables for both fruit and size columns:")
print(all_dummies)

all_dummies = pd.get_dummies(df[['fruit', 'size']],drop_first=True)
print("\nDummy variables for both fruit and size columns Drop First:")
print(all_dummies)

# To add the dummies back to the original dataframe
df_with_dummies = pd.concat([df, fruit_dummies], axis=1)
print("\nOriginal dataframe with dummies added:")
print(df_with_dummies)
```

```python
import pandas as pd

# Create a dataframe with an integer column
df = pd.DataFrame({
    'id': [1, 2, 3, 4, 5],
    'category_id': [1, 2, None, 3, 2],
    'value': [100, 200, 150, 250, 180]
})

print("Original dataframe:")
print(df)

price_dummies = pd.get_dummies(df['value'], prefix='price', drop_first=True, dtype=int)

print("\nDummy variables for the integer column 'value':")
print(price_dummies)

# Apply get_dummies to the integer column 'category_id'
category_dummies = pd.get_dummies(df['category_id'], prefix='category', dtype=int)

print("\nDummy variables for the integer column 'category_id':")
print(category_dummies)

# Add the dummies back to the original dataframe
df_with_dummies = pd.concat([df, category_dummies], axis=1)

print("\nOriginal dataframe with dummies added:")
print(df_with_dummies)
print(pd.concat([df, category_dummies], axis=0))
```
